```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import csv
```

```python
path='evaluation_data/dblp_acm'
```

```python
word_count_dict = {}
total_number_of_cell = 0
word_list = []
```

```python
csv_file = open(path + '/train.csv', "r")
reader = csv.reader(csv_file)
for item in reader:
    for attr in item:
        total_number_of_cell += 1
        temp_length = len(attr.split(' '))
        if temp_length in word_count_dict:
            word_count_dict[temp_length] += 1
        else:
            word_count_dict[temp_length] = 1
        word_list.append(temp_length)
csv_file.close()
```

```python
csv_file = open(path + '/validation.csv', "r")
reader = csv.reader(csv_file)
for item in reader:
    for attr in item:
        total_number_of_cell += 1
        temp_length = len(attr.split(' '))
        if temp_length in word_count_dict:
            word_count_dict[temp_length] += 1
        else:
            word_count_dict[temp_length] = 1
        word_list.append(temp_length)
csv_file.close()
```

```python
csv_file = open(path + '/test.csv', "r")
reader = csv.reader(csv_file)
for item in reader:
    for attr in item:
        total_number_of_cell += 1
        temp_length = len(attr.split(' '))
        if temp_length in word_count_dict:
            word_count_dict[temp_length] += 1
        else:
            word_count_dict[temp_length] = 1
        word_list.append(temp_length)
csv_file.close()
```

```python
plt.figure(dpi = 600)
plt.title('Sentence Length Distribution in DBLP-ACM Dataset')
plt.ylabel('Density')
plt.xlabel('Sentence Length')
sns.distplot(word_list,hist = True, norm_hist=False)
plt.show()
```
